```python
df=covid.groupby(by=['country','observation_date']).sum().reset_index()
df
```

```python
data=df.drop(['SNo'],axis=1)
data
```

```python
data=covid.groupby(by=['country','observation_date']).agg({'confirmed':'sum','deaths':'sum','recovered':'sum'}).sort_values(['country','observation_date'],ascending=[True,True]).reset_index()
data
```

```python
def monotonic_increase(x):
    if np.sum(np.diff(x)<0)>0:
        return sorted(x)
    else:
        return x
```

```python
monitor=data.groupby(by=['country']).transform(lambda x:monotonic_increase(x))
monitor
```

```python
data[monitor.columns]=monitor.values
```

```python
data
```

```python
data=data.groupby(by=['observation_date']).agg({'confirmed':'sum','deaths':'sum','recovered':'sum'}).reset_index()
```

```python
data.head()
```

```python
data['observation_date']=pd.to_datetime(data['observation_date'])
```

```python
data
```

```python
data['day_since']=(data['observation_date']-data['observation_date'].min()).dt.days
data.head()
```

## Modelling

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
x=data['day_since'].values.reshape(-1,1)
y=data['confirmed'].values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
print("intercept:",regressor.intercept_)
print('coef:',regressor.coef_)
```

```python
y_pred=regressor.predict(x_test)
```

```python
r2=metrics.r2_score(y_test,y_pred)
n=x_test.shape[0]#sample size(datanin numenelerinin olcusu)
p=x_test.shape[1]# number of predictors(columnlarin sayi)

print('R^2:' , metrics.r2_score(y_test,y_pred))
print('Adjusted R^2:',1-(1-r2)*(n-1)/(n-p-1))
print('Mean Absolute Error:' , metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error:' , metrics.mean_squared_error(y_test,y_pred)) 
print('Root Mean Squared Error:' , np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
